```python
target = 'cluster'
X = df.drop(['cluster', target], axis=1)
y = df[target]
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier

# Faqat numerical ustunlar uchun preprocessor
preprocessor = StandardScaler()

# Pipeline yaratish
pipeline = Pipeline([
    ('preprocess', preprocessor),
    ('regressor', RandomForestClassifier(random_state=42))
])
```

```python
pipeline.fit(X, y)
```

```python
from joblib import dump
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier

# Misol uchun pipeline (faqat numerical features uchun)
pipeline = Pipeline([
    ('preprocess', StandardScaler()),
    ('regressor', RandomForestClassifier(random_state=42))
])

# Saqlash yo'li
save_path = r"C:\Users\Rasulbek907\Desktop\Project_MP\Models\Pipeline_Models\RandomForestClassifier.joblib"

# Pipeline'ni saqlash
dump(pipeline, save_path)

print(f"Pipeline muvaffaqiyatli saqlandi: {save_path}")
```
